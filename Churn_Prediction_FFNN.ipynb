## Data Preprocessing

```python
import category_encoders as ce
from sklearn.experimental import enable_iterative_imputer  # Import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder
import numpy as np

def preprocess_data(df):
    # Create a copy of the DataFrame
    df_processed = limit_churn.copy()

    # Replace missing values using iterative imputation for numeric columns
    imputer_numeric = IterativeImputer()
    df_processed[df.select_dtypes(include=[np.number]).columns] = imputer_numeric.fit_transform(df.select_dtypes(include=[np.number]))

    # Bayesian encoding for categorical columns
    encoder = ce.TargetEncoder(cols=["MRG", "TENURE", "REGION", "TOP_PACK"])
    df_processed[["MRG", "TENURE", "REGION", "TOP_PACK"]] = encoder.fit_transform(df_processed[["MRG", "TENURE", "REGION", "TOP_PACK"]], df_processed["CHURN"])

    # Fill missing values with a new category for categorical columns
    for column in df.select_dtypes(include=["object"]).columns:
        df_processed[column].fillna("Missing", inplace=True)

    # Label encoding for categorical columns
    le = LabelEncoder()
    for column in df_processed.select_dtypes(include=["object"]).columns:
        df_processed[column] = le.fit_transform(df_processed[column])

    return df_processed

df = preprocess_data(limit_churn)
```

## Class Imbalance handling

```python
from sklearn.utils import resample
import pandas as pd

# Oversampling of the minority class
mask = df["CHURN"] == 1
churn_df = df[mask].sample(n=100000, replace=True, random_state=42)
not_churn_df = df[~mask].sample(n=100000, random_state=42)
df_upsampled = resample(churn_df, replace=True, n_samples=100000, random_state=42)
df_final = pd.concat([not_churn_df, df_upsampled])

```

## Split the data into train test and Validation

```python
X = df_final.drop(['user_id', 'CHURN'], axis = 1)
```

```python
y = df_final["CHURN"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Split data into train, validation, and test sets
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

# Split the dataset into train and remaining (val+test) sets
X_train, X_remaining, y_train, y_remaining = train_test_split(X, y, test_size=(val_ratio + test_ratio), random_state=42)

# Split the remaining data into validation and test sets
X_val, X_test, y_val, y_test = train_test_split(X_remaining, y_remaining, test_size=(test_ratio / (val_ratio + test_ratio)), random_state=42)
```

## First Feed forward Neural Network

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# Define a simple feed-forward neural network
class FNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(FNN, self).__init__()
        self.hidden = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.output = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.hidden(x)
        x = self.relu(x)
        x = self.output(x)
        return x 

input_size = X_train.shape[1]
hidden_size = 128
output_size = 2  # binary classification (churn vs. not churn)
model = FNN(input_size, hidden_size, output_size)
loss_function = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Convert the train, validation, and test data to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.to_numpy(), dtype=torch.long)
X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val.to_numpy(), dtype=torch.long)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.to_numpy(), dtype=torch.long)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

num_epochs = 3
for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader):
        # Forward pass
        output = model(images)
        loss = loss_function(output, labels)
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        # Update weights
        optimizer.step()

# Evaluation on validation set
val_predictions = model(X_val_tensor)
val_loss = loss_function(val_predictions, y_val_tensor)
val_accuracy = (val_predictions.argmax(dim=1) == y_val_tensor).sum().item() / len(y_val_tensor)

# Evaluation on test set
test_predictions = model(X_test_tensor)
test_loss = loss_function(test_predictions, y_test_tensor)
test_accuracy = (test_predictions.argmax(dim=1) == y_test_tensor).sum().item() / len(y_test_tensor)

print('Validation Loss: {:.4f}, Accuracy: {:.2f}%'.format(val_loss.item(), val_accuracy * 100))
print('Test Loss: {:.4f}, Accuracy: {:.2f}%'.format(test_loss.item(), test_accuracy * 100))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# After model prediction
test_predictions = model(X_test_tensor)
_, predicted = torch.max(test_predictions, 1)

# Convert tensors to numpy arrays
predicted = predicted.numpy()
y_test = y_test_tensor.numpy()

# Calculate metrics
accuracy = accuracy_score(y_test, predicted)
precision = precision_score(y_test, predicted)
recall = recall_score(y_test, predicted)
f1 = f1_score(y_test, predicted)

print('Accuracy: {:.2f}%'.format(accuracy * 100))
print('Precision: {:.2f}%'.format(precision * 100))
print('Recall: {:.2f}%'.format(recall * 100))
print('F1 Score: {:.2f}%'.format(f1 * 100))
```

## Second model using leakyRelu, learning rate of 0.1 and additional hidden layer


```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import recall_score

# Define a feed-forward neural network with two hidden layers
class FNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(FNN, self).__init__()
        self.hidden1 = nn.Linear(input_size, hidden_size)
        self.hidden2 = nn.Linear(hidden_size, hidden_size)
        self.leakyrelu = nn.LeakyReLU()
        self.output = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.hidden1(x)
        x = self.leakyrelu(x)
        x = self.hidden2(x)
        x = self.leakyrelu(x)
        x = self.output(x)
        return x 

input_size = X_train.shape[1]
hidden_size = 128
output_size = 2  # binary classification (churn vs. not churn)
model = FNN(input_size, hidden_size, output_size)
loss_function = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)  # Increase the learning rate

# Convert the train, validation, and test data to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.to_numpy(), dtype=torch.long)
X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val.to_numpy(), dtype=torch.long)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.to_numpy(), dtype=torch.long)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

num_epochs = 3
for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader):
        # Forward pass
        output = model(images)
        loss = loss_function(output, labels)
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        # Update weights
        optimizer.step()

# Evaluation on validation set
val_predictions = model(X_val_tensor)
val_loss = loss_function(val_predictions, y_val_tensor)
val_predictions_classes = val_predictions.argmax(dim=1)
val_accuracy = (val_predictions_classes == y_val_tensor).sum().item() / len(y_val_tensor)
val_recall = recall_score(y_val_tensor, val_predictions_classes)

# Evaluation on test set
test_predictions = model(X_test_tensor)
test_loss = loss_function(test_predictions, y_test_tensor)
test_predictions_classes = test_predictions.argmax(dim=1)
test_accuracy = (test_predictions_classes == y_test_tensor).sum().item() / len(y_test_tensor)
test_recall = recall_score(y_test_tensor, test_predictions_classes)

print('Validation Loss: {:.4f}, Accuracy: {:.2f}%, Recall: {:.2f}%'.format(val_loss.item(), val_accuracy * 100, val_recall * 100))
print('Test Loss: {:.4f}, Accuracy: {:.2f}%, Recall: {:.2f}%'.format(test_loss.item(), test_accuracy * 100, test_recall * 100))
```

## Third model using lightning

```python
import os
import torch
from torch import optim, nn
from torch.utils.data import DataLoader, TensorDataset
import pytorch_lightning as pl
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# Define the Feedforward Neural Network
class FNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(FNN, self).__init__()
        self.hidden1 = nn.Linear(input_size, hidden_size)
        self.hidden2 = nn.Linear(hidden_size, hidden_size)
        self.leakyrelu = nn.LeakyReLU()
        self.output = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.hidden1(x)
        x = self.leakyrelu(x)
        x = self.hidden2(x)
        x = self.leakyrelu(x)
        x = self.output(x)
        return x

# Define the LightningModule
class LitFNN(pl.LightningModule):
    def __init__(self, model):
        super().__init__()
        self.model = model

    def training_step(self, batch, batch_idx):
        images, labels = batch
        output = self.model(images)
        loss = nn.CrossEntropyLoss()(output, labels)
        self.log("train_loss", loss)
        return loss

    def validation_step(self, batch, batch_idx):
        images, labels = batch
        output = self.model(images)
        preds = output.argmax(dim=1)
        accuracy = accuracy_score(labels.cpu(), preds.cpu())
        recall = recall_score(labels.cpu(), preds.cpu())
        precision = precision_score(labels.cpu(), preds.cpu())
        f1 = f1_score(labels.cpu(), preds.cpu())
        metrics = {'val_accuracy': accuracy, 'val_recall': recall, 'val_precision': precision, 'val_f1': f1}
        self.log_dict(metrics)

    def configure_optimizers(self):
        optimizer = optim.Adam(self.model.parameters(), lr=1e-3)
        return optimizer

# Parameters
input_size = X_train.shape[1]
hidden_size = 128
output_size = 2  # binary classification (churn vs. not churn)

# Convert the train, validation, and test data to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.to_numpy(), dtype=torch.long)
X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val.to_numpy(), dtype=torch.long)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)

# Initialize the model and the LightningModule
fnn_model = FNN(input_size, hidden_size, output_size)
lit_fnn = LitFNN(fnn_model)

# Train the model
trainer = pl.Trainer(max_epochs=10)
trainer.fit(lit_fnn, train_loader, val_loader)
```

```python
class MetricsCallback(pl.Callback):
    def on_validation_end(self, trainer, pl_module):
        print(trainer.callback_metrics)

# Initialize the callback
metrics_callback = MetricsCallback()

# Pass the callback to the trainer
trainer = pl.Trainer(max_epochs=10, callbacks=[metrics_callback])
trainer.fit(lit_fnn, train_loader, val_loader)
```
